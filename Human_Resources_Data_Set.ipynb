```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "/content/HRDataset_v14.csv"
df = pd.read_csv(file_path)

# Set style for better visuals
sns.set_style("whitegrid")

# Display all columns in the DataFrame
pd.set_option('display.max_columns', None)

# --- 1. Histogram of Salaries ---
plt.figure(figsize=(10,5))
sns.histplot(df['Salary'], bins=20, kde=True)
plt.title("Salary Distribution", fontsize=14)
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()

# --- 2. Boxplot of Salary by Department ---
plt.figure(figsize=(12,6))
sns.boxplot(x=df['Department'], y=df['Salary'])
plt.title("Salary Distribution by Department", fontsize=14)
plt.xlabel("Department")
plt.ylabel("Salary")
plt.show()

# --- 3. Countplot of Employee Performance Scores ---
plt.figure(figsize=(8,5))
sns.countplot(x=df['PerfScoreID'])
plt.title("Employee Performance Scores", fontsize=14) # Corrected line: Added title
plt.xlabel("Performance Score ID") # Added x-axis label for clarity
plt.ylabel("Count") # Added y-axis label for clarity
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "/content/HRDataset_v14.csv"
df = pd.read_csv(file_path)

# Ensure "Absences" column exists (update if needed)
absence_column = "Absences"  # Adjust if your column name differs

# Set the style to match your reference image
sns.set_style("whitegrid")  # Light background for clarity

# --- Create Boxplot ---
plt.figure(figsize=(10,6))
ax = sns.boxplot(
    x="Department",
    y=absence_column,
    data=df,
    palette="Oranges",  # Matches your color theme
    flierprops={'marker': 'o', 'color': 'black', 'markersize': 7}  # Black dots for outliers
)

# --- Formatting Titles & Labels ---
plt.title("Absences\nBy Department", fontsize=16, fontweight='bold', ha='left', x=0.02, y=1.05)
plt.xlabel("Department", fontsize=12, fontweight="bold")
plt.ylabel("Absences", fontsize=12, fontweight="bold")
plt.xticks(rotation=30, ha="right")  # Rotate department names for readability
plt.yticks(fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.7)  # Light grid for readability

# Remove top and right border for cleaner look
sns.despine()

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = "/content/HRDataset_v14.csv"
df = pd.read_csv(file_path)

# Ensure "Gender" and "Department" columns exist
# The column name for gender is likely 'Sex', not 'Gender'
gender_column = "Sex"  # Correcting the column name
department_column = "Department"  # Adjust if needed

# Count occurrences of gender per department
gender_counts = df.groupby([department_column, gender_column]).size().unstack()

# Create stacked bar chart
fig, ax = plt.subplots(figsize=(12,6))
bars = gender_counts.plot(kind="bar", stacked=True, color=["blue", "red"], ax=ax)

# Add labels to each stack
for i, department in enumerate(gender_counts.index):
    bottom_value = 0  # Start at the bottom of the stack
    for gender in gender_counts.columns:
        count = gender_counts.loc[department, gender]
        if count > 0:  # Only label if there is a value
            ax.text(i, bottom_value + count / 2, str(int(count)), ha='center', va='center', fontsize=10, color='white', fontweight='bold')
        bottom_value += count  # Move up the stack for next label

# Formatting
plt.title("Gender Count by Department", fontsize=16, fontweight='bold')
plt.xlabel("Department", fontsize=12, fontweight='bold')
plt.ylabel("Count", fontsize=12, fontweight='bold')
plt.xticks(rotation=30, ha="right")
plt.legend(title="Gender", labels=["Men", "Women"]) # You might want to adjust legend labels based on the values in the 'Sex' column
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show plot
plt.show()
```
